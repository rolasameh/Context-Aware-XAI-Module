```python
import pandas as pd
import numpy as np

# Specify the folder path and filename
file_path = 'C:/Users/hp/Downloads/training_1m.csv'
file_path2 = 'C:/Users/hp/Downloads/Movies_Modified.csv'
file_encoding = 'ISO-8859-1'

# Read the file into a DataFrame
df2 = pd.read_csv(file_path2, encoding=file_encoding)
df = pd.read_csv(file_path)

# Filter the DataFrame to only include rows where the rating is 4 or higher
high_ratings = df[df['rating'] >= 4]

# Create an array of the seven days of the week
days_of_week = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

# Generate random day values for the number of rows in high_ratings DataFrame
random_days = np.random.choice(days_of_week, size=high_ratings.shape[0])

# Assign the random days to the 'Day' column in high_ratings DataFrame
high_ratings.loc[:, 'day'] = random_days

# Get the total number of rows in the filtered DataFrame
num_rows = high_ratings.shape[0]

# Print the total number of rows
print("Total number of rows:", num_rows)

# View the updated DataFrame with the 'Day' column
print(high_ratings.head(100))
```

```python
# Merge the DataFrames on the 'movieID' column
merged_df = pd.merge(high_ratings, df2, on='movieID')

num_rows = high_ratings.shape[0]

# Print the total number of rows
print("Total number of rows:", num_rows)
merged_df.head(20)
```

```python
# Loop over the unique values in the 'userID' column
for user_id in merged_df['userID'].unique():
    # Create a separate DataFrame for this user
    user_df = merged_df[merged_df['userID'] == user_id]

    # Count the occurrences of each genre for this user
    genre_counts = {}
    for row in user_df['genre']:
        genres = row.split('|')
        for genre in genres:
            if genre not in genre_counts:
                genre_counts[genre] = 0
            genre_counts[genre] += 1

    # Print the most watched genre for this user
    most_watched_genre = max(genre_counts, key=genre_counts.get)
    print(f"User {user_id}: Most watched genre is {most_watched_genre}")
```

```python

# Create a dictionary to store the most watched genre for each user
user_genre_dict = {}

# Loop over the unique values in the 'userID' column
for user_id in merged_df['userID'].unique():

    # Create a separate DataFrame for this user
    user_df = merged_df[merged_df['userID'] == user_id]

    # Loop over the unique values in the 'genre' column for this user
    genre_counts = {}
    for genre in user_df['genre'].unique():

        # Split any genres separated by '|' and count them individually
        genres = genre.split('|')
        for g in genres:
            if g not in genre_counts:
                genre_counts[g] = 0
            genre_counts[g] += user_df[user_df['genre'].str.contains(g)]['rating'].count()

    # Determine the most watched genre for this user
    most_watched_genre = max(genre_counts, key=genre_counts.get)
    
    # Add the most watched genre for this user to the dictionary
    user_genre_dict[user_id] = most_watched_genre

# Create a new DataFrame with the most watched genre for each user
preferred_genre_df = pd.DataFrame(list(user_genre_dict.items()), columns=['userID', 'preferred_genre'])
preferred_genre_df.head(20)
```
